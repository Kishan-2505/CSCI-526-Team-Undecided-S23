```python
import matplotlib.pyplot as plt

# sample data
data = json_response_level1['data']
# filter data
filtered_data = [d for d in data if d.get('causeOfDeath') == 3]

# calculate percentages
num_true = sum(1 for d in filtered_data if d.get('isGettingSmall') is True)
num_false = sum(1 for d in filtered_data if d.get('isGettingSmall') is False)
total = num_true + num_false
percent_true = num_true / total * 100
percent_false = num_false / total * 100

# plot pie chart
fig, ax = plt.subplots()
ax.pie([percent_true, percent_false], labels=['Shrinking', 'Infinite heaalth'], autopct='%1.1f%%')
ax.set_title('Percentage Winning of players Shrinking or have Infinite health ')
plt.show()
```

```python
# Filter out data with bulletsFired=0 using a lambda function
filtered_data = list(filter(lambda x: x['bulletsFired'] != 0))

# Create two lists for bulletsFired and bulletHit-bulletsFired values
bullets_fired = [x['bulletsFired'] for x in filtered_data]
bullet_hit_diff = [x['bulletHit'] - x['bulletsFired'] for x in filtered_data]

# Create a pie chart with labels and colors
plt.pie([sum(bullets_fired), sum(bullet_hit_diff)], labels=['Bullets Fired', 'Bullets Hit'])
```

```python
import matplotlib.pyplot as plt

# Filter out data with bulletsFired=0 using a lambda function
filtered_data = list(filter(lambda x: x['bulletsFired'] != 0, json_response_level2['data']))
# print(filtered_data)
# Create two lists for bulletsFired and bulletHit-bulletsFired values
s=0
for x in filtered_data:
    s+=x['bulletsFired']
# bullets_fired = [x['bulletsFired'] for x in filtered_data]
# bullet_hit_diff = [x['bulletHit'] - x['bulletsFired'] for x in filtered_data]
print(s)
t=0
for x in filtered_data:
    t+=x['bulletHit']
print(t)
# Calculate the total number of bullets hit and missed
total_bullets_hit = t
total_bullets_missed = s-t

# Create a pie chart with labels and colors
labels = ['Bullets Hit', 'Bullets Missed']
colors = ['#1f77b4', '#ff7f0e']
plt.pie([total_bullets_hit, total_bullets_missed], labels=labels, colors=colors, autopct='%1.1f%%')

# Add a title
plt.title('Bullet Hit Ratio')

# Display the chart
plt.show()
```
